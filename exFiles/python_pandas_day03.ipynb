```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import json as j

# 시각화 라이브러리 (내장 데이터셋 포함함)
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')  # filterwarnings를 무시=> 경고메세지 무시

print('numpy version , ', np.__version__)
print('pandas version , ', pd.__version__)

# 배열에 대한 정보를 쉽게 확인하기 위하여 함수를 만들어두면 편리하다
def aryInfo(ary) :
    print('type - ' , type(ary))
    print('shape - ', ary.shape)   
    print('ndim - ', ary.ndim)
    print('dtype - ', ary.dtype)
    print()
    print('data - ')
    print(ary)     # list 와의 차이점 : ',' 없음. 배열로 보면 됨!
    
    
# series에 대한 정보를 쉽게 확인하기 위하여 함수를 만들어두면 편리하다
def seriesInfo(s) :
    print('type - ' , type(s))
    print('index - ', s.index)
    print('values - ', s.values)   
    print('dtype - ', s.dtype)
    print()
    print('data - ')
    print(s)
    
# frame에 대한 정보 확인 함수
def frmInfo(frm) :
    print('type    - ' , type(frm))
    print('shape   - ', frm.shape)
    print('ndim    - ', frm.ndim)
    print('row idx - ', frm.index, type(frm.index))
    print('col idx - ', frm.columns , type(frm.columns))
    print('values - ', type(frm.values))
    print(frm.values)
    print('data - ')
    print(frm)
```

프레임 병합
- pd.merged(frm, frm, how) : 공통의 열을 기준으로 병합
- pd.join()   : 인덱스를 기준으로 병합
- concat()    : 연결


```python
# dict 형식으로 만들기 (key : value 로 이루어진 데이터셋)
frm01 = pd.DataFrame({
    '학번' : [1,2,3,4],
    '이름' : ['김', '오', '최', '서'],
    '학년' : [5, 6, 9, 6]
})
display(frm01)

# 두개 만들기 => 열인덱스가 동일해야함 (학번, 이름, 학년)
frm02 = pd.DataFrame({
    '학번' : [5,4,2,3],
    '학과' : ['건강', '건강', 'MIS', '무역'],
    '학점' : [3.9, 3.8, 3.7, 4.5]
})
display(frm02)
```

```python
pd.merge(frm01, frm02, how = 'inner')
# how의 default 는  inner로 생략해도 inner를 넣어준 것과 같은 결과 나옴
```

```python
pd.merge(frm01, frm02, how = 'left')   # 차집합 역할
```

```python
pd.merge(frm01, frm02, how = 'right')   # 차집합 역할
```

```python
# 위에서 copy해옴..
frm01 = pd.DataFrame({
    '학번' : [1,2,3,4],
    '이름' : ['김', '오', '최', '서'],
    '학년' : [5, 6, 9, 6]
})
display(frm01)

# 공통의 열이 없는 상태
frm02 = pd.DataFrame({   
    '과목코드' : [5,4,2,3], 
    '학과' : ['건강', '건강', 'MIS', '무역'],
    '학점' : [3.9, 3.8, 3.7, 4.5]
})
display(frm02)
```

```python
# pd.merge(frm01, frm02, how='inner')   # => 오류
pd.merge(frm01, frm02, how='inner', left_on = '학번', right_on = '과목코드')
```
