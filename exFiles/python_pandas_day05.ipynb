```python
titanicFrm = sns.load_dataset('titanic')
titanicFrm.head()
```

```python
print('Q1) 성별에 따른 선실등급의 인원 수를 데이터 프레임으로 만들고 싶다면? - ')
# 성별에 따른 선실등급별 인원 수
grp = titanicFrm.groupby(['sex', 'pclass']).size().reset_index(name = 'cnt')
grp
```

```python
# 성별에 따른 선실등급별 인원수 
grp.pivot(index = 'sex', columns = 'pclass', values = 'cnt')
```

```python
print('Q2) 성별과 생존여부에 따른 인원수를 데이터 프레임으로 만들고 싶다면?')
print('groupby - ')
grp = titanicFrm.groupby(['sex', 'survived']).size().reset_index(name = 'cnt')
print('pivot - ')
grp.pivot('sex', 'survived', 'cnt')
```

```python
titanicFrm['cnt'] = 1     # -> count를 위해 모든 값에 1 부여
# titanicFrm.head()
titanicFrm.pivot_table(index    = 'sex'
                       ,columns = 'survived'
                       ,values  = 'cnt'
                       ,aggfunc = [np.sum]  # 집계함수 수행 가능. 각 그룹 내에서 cnt 열의 값을 합산
                       ,margins = True 
                       ,margins_name = '분석결과')   # 성별에 따른 생존여부에 따른 집계
```

```python
print('Q1) 네개의 나이 그룹을 만들면?')
# age를 4개로 분류해서 라벨을 각각 부여
titanicFrm['age_category'] = pd.qcut(titanicFrm['age'], 4, labels = ['청소년','청년','중년','장년'])
```

```python
# 어떤 값 있는지 확인
titanicFrm['age_category'].unique()
```

```python
# 생존자들의 비율
# survive = titanicFrm['survived'] / titanicFrm['cnt'].count()
# display(survive)

subset = titanicFrm.loc[ : , ['sex','pclass','age_category', 'survived']]
subset.head()
```

```python
# 인덱스는 성별과 나이 그룹에 대한 다중인덱스 사용하고 열인덱스는 선실그룹  # => 멀티인덱스 사용
resultFrm = subset.pivot_table(index = ['sex', 'age_category'],
                  columns = 'pclass',
                  aggfunc = 'sum',    # => count로 하면 X
                  margins = True,      # => 전체 인원수 구해줌
                  margins_name = '생존율')    
```

```python
# 생존율을 데이터로 사용(생존인원수/전체인원수)
print('pivot 생성 - ')
print('index - ' ,resultFrm.index)    # index 확인
print('columns - ' ,resultFrm.columns)   # columns 확인
# resultFrm.loc[('생존율', '') ,('survived','생존율')]   # => 714
resultFrm / resultFrm.loc[('생존율', '') ,('survived','생존율')]
```
