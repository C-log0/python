```python
# 배열에 대한 정보를 쉽게 확인하기 위하여 함수를 만들어두면 편리하다
def aryInfo(ary) :
    print('type - ' , type(ary))
    print('shape - ', ary.shape)   
    print('ndim - ', ary.ndim)
    print('dtype - ', ary.dtype)
    print()
    print('data - ')
    print(ary)     # list 와의 차이점 : ',' 없음. 배열로 보면 됨!
    
    
# series에 대한 정보를 쉽게 확인하기 위하여 함수를 만들어두면 편리하다
def seriesInfo(s) :
    print('type - ' , type(s))
    print('index - ', s.index)
    print('values - ', s.values)   
    print('dtype - ', s.dtype)
    print()
    print('data - ')
    print(s)
    
# frame에 대한 정보 확인 함수
def frmInfo(frm) :
    print('type    - ' , type(frm))
    print('shape   - ', frm.shape)
    print('ndim    - ', frm.ndim)
    print('row idx - ', frm.index, type(frm.index))
    print('col idx - ', frm.columns , type(frm.columns))
    print('values - ', type(frm.values))
    print(frm.values)
    print('data - ')
    print(frm)
```

```python
# 파생 컬럼(변수) 만들기
print('Q1) age_by_10 컬럼을 추가하고싶다(기존 나이에 10살 더한 값)')
# 나이는 실수형일수 없으므로 int 로 형변환 -> 불가 (typeError) -> Series 이므로
# titanicFrm['age_by_10'] = int(titanicFrm['age'] + 10)
titanicFrm['age_by_10'] =( (titanicFrm['age'] + 10).values ).astype('int')
titanicFrm.head()
```

```python
print('Q2) parch와 sibsp를 더한 값에 1을 추가해서 family_no 컬럼 추가')
titanicFrm['family_no'] = titanicFrm['parch'] + titanicFrm['sibsp'] + 1
titanicFrm.head()
```

```python
print('Q3) 방금 만든 2개의 컬럼을 삭제하고 싶다면 - ')
print('drop( , axis = 1 )')
titanicFrm.drop(['age_by_10','family_no'],
                axis = 1,
                inplace = True)
```

```python
print('Q4) 요금(fare)에 대한 통계(최대,최소,평균,합계) 확인이 필요하다 - ')
# titanicFrm['fare'].describe()   # => 문자열이 아닌 전체 정보 확인
print('sum - ', np.sum(titanicFrm['fare'].values))  # => 배열을 꺼내 계산
```
