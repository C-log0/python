#### Quiz
- zipcode.txt
- input() 함수를 이용해서 동 이름을 입력   예)
- 예) 개포 (startswith)
- 해당하는 동의 주소를 출력하는 함수 (searchAddr()) 를 정의

```python
def searchAddr() :
    dong = input('동을 입력하세요 : ')
    print(dong)
    with open('../data/zipcode.txt', 'r', encoding = 'utf-8') as file :
        addr = file.readline()
#         print('addr - ', addr)
#         print('type - ', type(addr))
        
        while addr :
            addr = addr.split()    # split('\t') 으로 해도 됨
    #         print('addr - ', addr)
    #         print('type - ', type(addr))
            if addr[3].startswith(dong) :
                print(addr)
            addr = file.readline() 
```
