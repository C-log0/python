```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import json as j

# 시각화 라이브러리 (내장 데이터셋 포함함)
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')  # filterwarnings를 무시=> 경고메세지 무시

print('numpy version , ', np.__version__)
print('pandas version , ', pd.__version__)

# 배열에 대한 정보를 쉽게 확인하기 위하여 함수를 만들어두면 편리하다
def aryInfo(ary) :
    print('type - ' , type(ary))
    print('shape - ', ary.shape)   
    print('ndim - ', ary.ndim)
    print('dtype - ', ary.dtype)
    print()
    print('data - ')
    print(ary)     # list 와의 차이점 : ',' 없음. 배열로 보면 됨!
    
    
# series에 대한 정보를 쉽게 확인하기 위하여 함수를 만들어두면 편리하다
def seriesInfo(s) :
    print('type - ' , type(s))
    print('index - ', s.index)
    print('values - ', s.values)   
    print('dtype - ', s.dtype)
    print()
    print('data - ')
    print(s)
    
# frame에 대한 정보 확인 함수
def frmInfo(frm) :
    print('type    - ' , type(frm))
    print('shape   - ', frm.shape)
    print('ndim    - ', frm.ndim)
    print('row idx - ', frm.index, type(frm.index))
    print('col idx - ', frm.columns , type(frm.columns))
    print('values - ', type(frm.values))
    print(frm.values)
    print('data - ')
    print(frm)
```

merge : 열병합
- (행)index를 기준으로 병합 가능
- how, left_on, right_on, left_index, right_index

```python
frm01 = pd.DataFrame({
    'cols01' : ['busan', 'seoul','jeju'] ,
    'cols02' : [1,2,3] ,
    'cols03' : [4,5,6] 
})
display(frm01)
print()
# 인자가 들어오는 순서는 (data,index,columns)지만, 이름을 적어주면 순서상관없음
frm02 = pd.DataFrame( data = np.arange(12).reshape(6,2) ,
                     columns = ['feature01', 'feature02']
                    )   
display(frm02)
```

```python
pd.merge(frm01, frm02, right_index = True, left_index = True)
```

```python
pd.merge(frm01, frm02, left_index = True, right_on = 'feature01')
# => ValueError
```

```python
pd.merge(frm01, frm02, right_index = True, left_on = 'cols02')
```

- join : 인덱스 병합의 프레임 전용 함수

```python
frm01.join(frm02, how = 'inner')
```

- pd.concat() : 인덱스나 열 인덱스를 기준으로 하지 않고 단순 데이터 연결
- 축의 방향이 필요하다 axis = 0 (행이 늘어난다)


```python
pd.concat( [pd.Series([8,9]) , pd.Series([4,5,6,7])] )
```

```python
pd.concat( [frm01, frm02], axis = 0 )
```

데이터를 분석하다보면 그룹의 통계량 확인이 필요한 경우가 있다.
- groupby (열인덱스 | 인덱스 )
- 데이터를 그룹으로 분할하여 독립된 그룹에 대해서 별도의 데이터 처리 또는 통계량
- split : 특징에 따라서 분류
- apply : 함수를 적용(집계, 필터링, 변환)
- combine : 처리된 결과를 하나의 프레임으로 결합
- agg([count, sum, mean, median, var, std, min, max...]) : 통계함수 ( ※ agg() : 집계함수 )

```python
frm = pd.read_csv('../../data/service_data_groupby_sample.csv',
                 encoding = 'cp949')
frm.head()
```

```python
print('Q1) 나이에 대한 평균을 확인하고 싶다면? - ')
# print(frm['age'].mean())
print( np.mean(frm['age']) )
```

```python
print('Q2) 지역별 나이평균을 확인하고 싶다면? -')
# split 이용 , 함수적용 , group by 이용 => combine
# frm.groupby('region')   # group 생성 (프레임 생성)
# frm.groupby('region').get_group('경기')
# np.mean(frm.groupby('region').get_group('경기')['age'])
frm.groupby('region')['age'].mean()
```

```python
print('Q3) 성별을 기준으로 그룹을 나누고 싶다면 - ')
# frm.groupby('gender').get_group('여자')
frm.groupby('gender')['height'].mean()
```

```python
frm.reset_index(inplace = True)
frm
```

```python
print('다중 통계량의 정보를 확인하고자 할 때 - agg() ')
# frm.groupby('gender').mean()    # gender로 지정된 모든값의 평균 구함
frm.groupby('gender').agg(['mean', 'var', 'std'])
```

```python
# index 이름 설정
# frm을 성별에 따라 height로 그룹화 하여 가지고 온다.
# '**'는 딕셔너리{}를 키워드 인수로 전달하는 방식이다.
# 'mean'을 'mean_height'로 인덱스명을 변경함
frm.groupby('gender')['height'].agg(**{'mean_height' : 'mean'}).reset_index()
```

```python
frm.groupby('gender')['height'].agg({'max','min', 'mean','median'}).reset_index()
```

```python
print('Q4) 성별에 따른 거주지의 최빈값을 확인하고 싶다면?')
frm.groupby('gender')['region'].agg(lambda x : x.mode() )   # => split 됨
```

```python
# 그룹의 기준이 하나 이상(멀티) => [] 사용
# frm.groupby(['gender','region']).get_group(('남자','서울')).mean() # 튜플 형식
tmp = frm.groupby(['gender','region'])['height'].agg(**{'m' : 'mean'})
```

```python
tmp.index # 그룹의 형식 : 튜플
```

```python
print('Q5) 성별에 따른 키의 최소값과 최대값을 구한다면?')
frm.groupby('gender')['height'].agg(['min', 'max'])
```

```python
# 다중 인덱스
frm.groupby('gender')[['height','age']].agg(['min', 'max'])  
```
